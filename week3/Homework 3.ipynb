```python
import pandas as pd
import altair as alt
df = pd.read_csv('Udemy.csv')
df1 = df[df['num_subscribers']>=10000]
df1 = df.sort_values(['num_subscribers','avg_rating','num_reviews'],ascending=False)
class_name = df1.iloc[0]['title']
class_rate = float(df1.iloc[0]['avg_rating'])
class_sub = df1.iloc[0]['num_subscribers']
print("The class with most subscribers is '{}' which has {} subscribers and {:.1f}/5 rate.".format(class_name,class_sub,class_rate))
```

```python
df2 = df[df['num_subscribers']>=10000]
df2 = df.sort_values(['avg_rating','num_subscribers','num_reviews'],ascending=False)
class_name = df2.iloc[0]['title']
class_rate = float(df2.iloc[0]['avg_rating'])
class_sub = df2.iloc[0]['num_subscribers']
print("The class with highest rate is '{}' which has {} subscribers and {:.1f}/5 rate.".format(class_name,class_sub,class_rate))
```

```python
df3 = df[df['num_subscribers']>=10000]
df3 = df.sort_values(['num_reviews','avg_rating','num_subscribers'],ascending=False)
class_name = df3.iloc[0]['title']
class_rate = float(df3.iloc[0]['avg_rating'])
class_sub = df2.iloc[0]['num_subscribers']
print("The class with most reviews is '{}' which has {} subscribers and {:.1f}/5 rate.".format(class_name,class_sub,class_rate))
```

```python
df = df1
alt.Chart(df).mark_circle().encode(
    x = 'avg_rating',
    y = 'num_subscribers',
    color = alt.Color('num_reviews', scale=alt.Scale(scheme='turbo')),
    size = 'num_reviews',
    opacity=alt.value(0.7),
    tooltip = ['title','num_subscribers','avg_rating']
)
```

```python
df['published_time_new'] = pd.to_datetime(df['published_time'])
alt.Chart(df).mark_circle().encode(
    x = 'published_time_new:T',
    y = 'num_subscribers',
    color = alt.Color('num_reviews', scale=alt.Scale(scheme='turbo')),
    size ='num_reviews',
    tooltip = ['title','num_subscribers','published_time_new']
)
```
