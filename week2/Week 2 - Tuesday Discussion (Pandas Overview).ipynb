## Working with columns

```python
df = sns.load_dataset("mpg")
```

```python
df["weight"]
```

```python
# This is an example of a pandas Series, 
# which is the second most important datatype in pandas
type(df["weight"])
```

```python
# Saving some keystrokes, but doesn't work if there is a space in the name
# there are also some other restrictions, like it can't start with a number
df.weight
```

```python
df["weight"] > 3500
```

```python
# Make a new column (it goes all the way on the right side)
df["above 3500"] = (df.weight > 3500)
df
```

```python
# If you later want to change the dataframe, it's often important to copy
df2 = df[["cylinders","acceleration"]]
df2
```

```python
df2["cylinders"] = 3
```

```python
# Here is the same thing using copy:
df3 = df[["cylinders","acceleration"]].copy()
df3
```

```python
# No warning this time
df3["cylinders"] = 3
df3
```

## Applying functions

```python
df = pd.DataFrame(rng.integers(0,4,size=(5,3)))
df
```

```python
# Apply a function to every entry:
df.applymap(np.cos)
```

```python
# Apply a function to each column, one at a time
df.apply(sum)
```

```python
# The same thing
df.apply(sum, axis=0)
```

```python
# Apply a function to each row, one at a time
df.apply(lambda row: row[0] + 2*row[1] + 3, axis=1)
```

## Accessing elements

```python
df = sns.load_dataset("mpg")
```

Use loc with the index names and column names:

```python
df.loc[25,"name"]
```

```python
# This is one of the only cases in Python where the right endpoint is included.
df.loc[25:30, "model_year":"name"]
```

Use iloc with positions.  (Often the index position is the same as the index name.)

```python
# Here we're back to not including the right endpoints.
df.iloc[25:30, 3:5]
```

## Getting a subset of the rows

```python
df = pd.DataFrame(rng.integers(0,100,size=(10,3)))
df
```

```python
# Get all the rows where the 1-entry is strictly greater than the 2-entry
df[df[1] > df[2]]
```

## Logic in pandas
(Warning: it's different from standard Python.)

```python
df = pd.DataFrame(rng.integers(0,10,size=(20,3)))
df
```

```python
df[(df[0] > 5) & (df[1] < 5)]
```

```python
df[(df[0] == 7) | (df[1] == 2)]
```

```python
# Negation
df[~df[1].isin(range(7))]
```

## Finding missing values

```python
df = pd.read_csv("../../data/FastFoodRestaurants.csv")
```

```python
df[df.websites.isna()]
```

```python
# Two ways to get the negation:
df[~df.websites.isna()]
```

```python
# The second to get the negation:
df[df.websites.notna()]
```
