```python
numeric_cols = [c for c in df.columns if is_numeric_dtype(df[c])]
```

```python
df2 = df[numeric_cols].copy()
```

```python
df2.head(10)
```

```python
df2["Energy"].isna()
```

```python
df2["Energy"].notna()
```

```python
df[df2["Energy"].isna()]
```

```python
df3 = df2[~df2["Energy"].isna()].copy()
```

```python
from sklearn.cluster import KMeans
```

```python
KMeans?
```

```python
kmeans = KMeans(6)
```

```python
kmeans.fit(df3)
```

```python
kmeans.predict(df3)
```

```python
df3["cluster"] = kmeans.predict(df3)
```

```python
df3.head(3)
```

```python
alt.Chart(df3).mark_circle().encode(
    x = "Energy",
    y = "Popularity",
    color = "cluster:O"
)
```

```python
alt.Chart(df3).mark_circle().encode(
    x = "Energy",
    y = "Popularity",
    color = "cluster:N"
)
```

```python
numeric_cols
```

```python
alt.Chart(df3).mark_point().encode(
    x = "Energy",
    y = "Popularity",
    size = "Loudness",
    shape = "cluster:N",
    color = "cluster:N"
)
```

```python
kmeans_new = KMeans(n_clusters=2)
```

```python
kmeans_new.fit(df3)
```

```python
df3["new_cluster"] = kmeans_new.predict(df3)
```

```python
df3.head()
```

```python
alt.Chart(df3).mark_circle().encode(
    x = "Artist Followers",
    y = "Popularity",
    color = "new_cluster:N"
)
```
