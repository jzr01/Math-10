```python
import altair as alt
import pandas as pd
from pandas.api.types import is_numeric_dtype
from vega_datasets import data
from sklearn.preprocessing import StandardScaler
```

```python
source = data.cars()
```

```python
col_a = "Weight_in_lbs"
col_b = "Miles_per_Gallon"
```

```python
df = source[[col_a,col_b,"Origin"]].copy()
```

```python
alt.Chart(df).mark_circle().encode(
    x = alt.X(col_a,scale=alt.Scale(zero=False)),
    y = alt.Y(col_b,scale=alt.Scale(zero=False)),
    color='Origin'
)
```

```python
#1
my_list = list(range(len(df)))
for i in range(len(df)):
    try: 
        int(df.iloc[i,0])
        int(df.iloc[i,1])
    except:
        my_list.remove(i)
df = df.iloc[my_list,:]
df
```

2.The dataset is natural to use sclaer because the number of two feature has huge difference.

```python
scaler = StandardScaler()
df2 = df[[col_a,col_b]]
scaler.fit(df2)
df2 = scaler.transform(df2)
```

```python
#3
df[[col_a,col_b]] = df2
```
